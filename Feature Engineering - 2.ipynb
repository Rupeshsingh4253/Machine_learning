```python
Q1. What is Min-Max scaling, and how is it used in data preprocessing? Provide an example to illustrate its application.

Min-Max scaling is a data preprocessing technique that transforms the values of a numeric feature to a specific range,
typically between 0 and 1. It is done by subtracting the minimum value of the feature and dividing by the range
(difference between maximum and minimum values)
Example:
Suppose you have a feature representing the salary of individuals with values ranging from $40,000 to $100,000. After Min-Max
scaling, these values could be transformed to a range between 0 and 1, making it easier for algorithms to converge and
improving model performance.

Q2. What is the Unit Vector technique in feature scaling, and how does it differ from Min-Max scaling? Provide an example to
illustrate its application.

Unit Vector scaling (also known as vector normalization) is a technique that scales the values of a feature to have a unit
norm, transforming the feature vector into a unit vector. Each data point is scaled by dividing it by the Euclidean norm of
the feature vector

Unlike Min-Max scaling, Unit Vector scaling does not set specific bounds for the feature values, but it ensures that the
magnitude of the feature vector is 1.

Example:
If you have a dataset with features representing different physical quantities (e.g., velocity, acceleration) and you want
to normalize them to unit vectors, you would apply the Unit Vector technique.

Q3. What is PCA (Principal Component Analysis), and how is it used in dimensionality reduction? Provide an example to
illustrate its application.

PCA (Principal Component Analysis) is a dimensionality reduction technique used to transform high-dimensional data into a
lower-dimensional space while retaining most of the original variance. It achieves this by identifying the principal
components, which are linear combinations of the original features.

Steps for PCA:

Standardize the data (optional but recommended).
Compute the covariance matrix.
Compute the eigenvectors and eigenvalues of the covariance matrix.
Sort the eigenvalues in descending order and choose the top 
�
k eigenvectors corresponding to the 
�
k largest eigenvalues.
Form a new matrix using the selected eigenvectors as columns.
Example:
Suppose you have a dataset with features representing different aspects of a person's health (e.g., heart rate, cholesterol levels, blood pressure). 
Applying PCA could help reduce these features to a smaller set of principal components capturing the most significant 
variability in the data.

Q4. What is the relationship between PCA and Feature Extraction, and how can PCA be used for Feature Extraction? Provide an 
example to illustrate this concept.

PCA and Feature Extraction:
PCA is a technique that can be used for feature extraction. In the context of PCA, features are transformed into a new set of
uncorrelated variables called principal components. These principal components capture the most significant information in the
original features, allowing for dimensionality reduction.

Example:
Consider a dataset with various demographic features (age, income, education level) and their correlation. Applying PCA could
yield principal components that represent combinations of these features. The first few principal components might capture the
majority of the variance in the data, serving as effective features for modeling.

Q5. You are working on a project to build a recommendation system for a food delivery service. The dataset contains features
such as price, rating, and delivery time. Explain how you would use Min-Max scaling to preprocess the data.

In the recommendation system project for a food delivery service:

Identify Numeric Features:

Identify the numeric features in the dataset, such as price, rating, and delivery time.
Apply Min-Max Scaling:

For each numeric feature, apply Min-Max scaling to transform the values to a standardized range (e.g., 0 to 1).
This ensures that features with different scales contribute equally to the recommendation model.
Example:

If the original price range is $10 to $50, Min-Max scaling would transform it to a range between 0 and 1. For instance, a food 
item priced at $30 might be scaled to 0.5, making it comparable to other features in the dataset.
Q6. You are working on a project to build a model to predict stock prices. The dataset contains many features, such as company
financial data and market trends. Explain how you would use PCA to reduce the dimensionality of the dataset.

For stock price prediction:

Standardize the Data:

Standardize the dataset to ensure that all features have a mean of 0 and a standard deviation of 1.
Apply PCA:

Apply PCA to the standardized dataset to obtain the principal components.
Choose the number of principal components based on the desired level of dimensionality reduction.
Example:

If the original dataset had features like company revenue, earnings per share, and market sentiment scores, PCA would identify
a reduced set of principal components that capture the most significant variations in the stock data. These components could be
used as input features for the prediction model.



```
