```python
Q1. A time series is a sequence of data points collected or recorded at successive equally spaced time intervals.
Time series analysis involves examining and analyzing data points collected over time to identify patterns, trends, 
and underlying structures. Common applications of time series analysis include:

- Economic forecasting: Predicting financial indicators such as stock prices, GDP growth, and inflation rates.
- Demand forecasting: Forecasting product demand for inventory management and production planning.
- Weather forecasting: Predicting future weather conditions based on historical data and meteorological models.
- Signal processing: Analyzing signals from sensors, instruments, or communication systems to extract useful information or 
    detect anomalies.
- Health monitoring: Analyzing medical data such as heart rate, blood pressure, and glucose levels for disease diagnosis and
    patient monitoring.
- Energy consumption forecasting: Predicting future energy consumption patterns for efficient resource allocation and
    energy management.

Q2. Some common time series patterns include:

- Trend: A long-term upward or downward movement in the data over time, indicating overall growth or decline.
- Seasonality: Repeating patterns or fluctuations in the data that occur at regular intervals, such as daily, weekly, monthly, 
    or yearly cycles.
- Cyclicality: Patterns or fluctuations in the data that occur at irregular intervals and are not necessarily related to
    seasonal effects.
- Randomness: Unpredictable or irregular variations in the data that do not follow any specific pattern.

These patterns can be identified and interpreted using techniques such as visual inspection of plots, decomposition of time
series into trend, seasonal, and residual components, and statistical tests for detecting periodicity and autocorrelation.

Q3. Time series data can be preprocessed before analysis by:

- Handling missing values: Imputing or interpolating missing data points to ensure continuity in the time series.
- Removing outliers: Identifying and removing extreme values that may distort the analysis or forecasting results.
- Resampling: Aggregating or disaggregating data points to adjust the time intervals or align with a specific frequency.
- Detrending and deseasonalizing: Removing trend and seasonal components to focus on the underlying patterns and fluctuations.
- Normalizing or standardizing: Scaling the data to a common range or distribution to facilitate comparison and interpretation.

Q4. Time series forecasting can be used in business decision-making to predict future trends, patterns, and behaviors based on 
historical data. Common applications include forecasting sales, demand, inventory levels, and financial metrics to support 
strategic planning, resource allocation, and risk management. However, some challenges and limitations of time series
forecasting include:

- Uncertainty: Future events and external factors may impact the accuracy of forecasts, leading to uncertainty and risk.
- Data quality: Forecasting accuracy depends on the quality, consistency, and completeness of historical data.
- Model complexity: Selecting and tuning appropriate forecasting models requires expertise and domain knowledge.
- Changing patterns: Time series data may exhibit evolving patterns and dynamics that are difficult to capture with static
    models.
- Overfitting: Overly complex models may fit the training data too closely and fail to generalize well to new data.

Q5. ARIMA (Autoregressive Integrated Moving Average) modelling is a popular approach for forecasting time series data. 
It combines autoregression (AR), differencing (I), and moving average (MA) components to capture the linear dependencies
and temporal structures present in the data. ARIMA models are widely used for univariate time series forecasting, especially 
when the data exhibit trend and seasonality patterns.

Q6. Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) plots help in identifying the order of ARIMA 
models by visualizing the correlation between a time series and its lagged values. ACF plots show the correlation between a
time series and its lagged values at various lags, while PACF plots show the correlation between a time series and its lagged 
values after removing the effects of shorter lags. The patterns observed in these plots can provide insights into the 
appropriate lag orders for the AR and MA components of an ARIMA model.

Q7. The assumptions of ARIMA models include:

- Stationarity: The time series data should exhibit constant mean, variance, and autocovariance over time.
- Linearity: The relationship between the time series and its lagged values should be linear.
- Absence of autocorrelation: The residuals (errors) of the model should be uncorrelated over time.

These assumptions can be tested using statistical tests, diagnostic plots, and residual analysis techniques to assess the
adequacy of the ARIMA model.

Q8. For monthly sales data for a retail store over the past three years, a seasonal ARIMA (SARIMA) model may be recommended
for forecasting future sales. SARIMA models are well-suited for time series data with trend and seasonality patterns, allowing
for the incorporation of seasonal components and capturing the seasonal variations in sales data effectively.

Q9. Some limitations of time series analysis include:

- Sensitivity to model assumptions: Time series models rely on assumptions such as stationarity, linearity, and independence 
    of residuals, which may not always hold true in practice.
- Limited predictability: Future events
```
