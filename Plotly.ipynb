````python
Q1: Scatter Plot for Age and Fare in Titanic Dataset
import seaborn as sns
import plotly.express as px

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# Plot scatter plot using Plotly express
fig = px.scatter(titanic, x='age', y='fare', title='Scatter Plot: Age vs Fare')
fig.show()
```

Q2: Box Plot for Tips Dataset
import plotly.express as px

# Load Tips dataset
tips = px.data.tips()

# Plot box plot using Plotly express
fig = px.box(tips, x='day', y='total_bill', title='Box Plot: Total Bill by Day')
fig.show()
```

Q3: Histogram with Color and Shape in Tips Dataset
import plotly.express as px

# Load Tips dataset
tips = px.data.tips()

# Plot histogram using Plotly express
fig = px.histogram(tips, x='sex', y='total_bill', color='day', pattern_shape='smoker',
                   title='Histogram: Total Bill by Sex with Smoker Pattern and Day Color')
fig.show()
```

Q4: Scatter Matrix Plot for Iris Datas
import plotly.express as px
# Load Iris dataset
iris = px.data.iris()
# Plot scatter matrix using Plotly express
fig = px.scatter_matrix(iris, dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                        color='species', title='Scatter Matrix Plot: Iris Dataset')
fig.show()

Q5: Distplot using Plotly express
import plotly.express as px

# Generate random data for demonstration
import numpy as np
data = np.random.randn(1000)

# Plot distplot using Plotly express
fig = px.histogram(data, nbins=30, marginal="rug", title='Distplot: Random Data')
fig.show()

Note: Distplot is a deprecated function in Plotly Express. Instead, you can use `px.histogram` as shown in the example.
````
