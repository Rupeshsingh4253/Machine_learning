```python
Q1. What is data encoding? How is it useful in data science?

Data encoding is the process of converting categorical or textual data into a numerical format that can be easily understood
and processed by machine learning algorithms. In data science, numerical representations are crucial because many machine 
learning models require numerical inputs. Data encoding is essential for handling categorical variables and ensuring that they 
can be effectively utilized in various algorithms.

Usefulness in Data Science:

Algorithm Compatibility: Many machine learning algorithms, such as linear regression or support vector machines, require 
numerical input features. Encoding allows us to represent categorical information in a way that is compatible with these 
algorithms.

Improved Model Performance: Proper encoding helps in preventing biases and misinterpretations that may arise if categorical
data is not appropriately transformed. It contributes to the overall performance and accuracy of machine learning models.

Facilitates Analysis: Numerical representations of categorical data make it easier to perform statistical analysis, visualize
relationships, and extract meaningful insights from the data.

Q2. What is nominal encoding? Provide an example of how you would use it in a real-world scenario.

Nominal encoding involves assigning unique numerical values to different categories without any specific order or ranking.
One common approach is using integer labels for each category. For example, assigning integers 1, 2, 3, and so on to different
categories.

Example:
Consider a dataset with a "Color" column containing categories like "Red," "Green," and "Blue." Nominal encoding could be 
applied as follows:

Red: 1
Green: 2
Blue: 3
Now, the categorical variable "Color" is numerically encoded in a nominal manner.

Q3. In what situations is nominal encoding preferred over one-hot encoding? Provide a practical example.

Nominal encoding is preferred over one-hot encoding when the categorical variable does not have a meaningful ordinal 
relationship, and the number of unique categories is relatively high. One-hot encoding creates binary columns for each 
category, leading to a high-dimensional and sparse representation.

Practical Example:
Consider a dataset with a "Country" column. One-hot encoding this column would result in many binary columns, each representing
a specific country. If there are numerous unique countries, the resulting dataset could become highly sparse and computationall
y expensive. Nominal encoding, where each country is assigned a unique integer, would be a more compact representation in such
cases.

Q4. Suppose you have a dataset containing categorical data with 5 unique values. Which encoding technique would you use to 
transform this data into a format suitable for machine learning algorithms? Explain why you made this choice.

For a dataset with 5 unique values, I would choose nominal encoding. Nominal encoding assigns a unique integer label to each
category without implying any ordinal relationship. This approach is suitable when the categories are not inherently ordered, 
and the algorithm needs a compact representation of the categorical variable.

Q5. In a machine learning project, you have a dataset with 1000 rows and 5 columns. Two of the columns are categorical, and
the remaining three columns are numerical. If you were to use nominal encoding to transform the categorical data, how many 

new columns would be created? Show your calculations.

If you use nominal encoding for categorical variables, each unique category is assigned a unique integer label. The number
of new columns created is equal to the number of categorical columns.

In this scenario, you have 2 categorical columns, so the number of new columns created would be 2.

Q6. You are working with a dataset containing information about different types of animals, including their species, habitat,
and diet. Which encoding technique would you use to transform the categorical data into a format suitable for machine learning
algorithms? Justify your answer.

For the dataset containing information about animal species, habitat, and diet, I would use nominal encoding. This is because
the categories like species, habitat, and diet are unlikely to have a meaningful ordinal relationship. Nominal encoding assigns
unique integer labels to each category without introducing any artificial ordering, making it suitable for such scenarios
. It provides a compact numerical representation of the categorical variables, which can be easily utilized by machine learning
algorithms.

Q7.You are working on a project that involves predicting customer churn for a telecommunications
company. You have a dataset with 5 features, including the customer's gender, age, contract type,
monthly charges, and tenure. Which encoding technique(s) would you use to transform the categorical
data into numerical data? Provide a step-by-step explanation of how you would implement the encoding.

For the given dataset with features like gender, contract type, and other numerical features, I would use a combination of
encoding techniques to transform the categorical data into numerical data. Here's a step-by-step explanation:

Features:

Gender (Categorical)
Contract Type (Categorical)
Age (Numerical)
Monthly Charges (Numerical)
Tenure (Numerical)
Encoding Techniques:

Nominal Encoding for Gender and Contract Type:

Use nominal encoding for gender and contract type since there is no inherent order or ranking in these categories.

Nominal encoding assigns a unique integer label to each category.

For Gender:
Male: 0
Female: 1
For Contract Type:
Month-to-month: 0
One year: 1
Two years: 2

| Gender | Contract Type | Age | Monthly Charges | Tenure |
|--------|---------------|-----|------------------|--------|
|   0    |       1       |  25 |      50.0        |   12   |
|   1    |       0       |  35 |      70.0        |   24   |
|   0    |       2       |  45 |      90.0        |   36   |
|   1    |       0       |  30 |      60.0        |   18   |
|   1    |       1       |  28 |      65.0        |   20   |
  
```
